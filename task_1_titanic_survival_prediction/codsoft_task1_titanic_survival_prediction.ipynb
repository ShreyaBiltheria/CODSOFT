## Plotting BoxPlot and Checking for Outliers

```python
# Selecting only numeric columns from the dataset
numeric_columns = titanic.select_dtypes(include='number')

# Creating a boxplot for numeric columns
sns.boxplot(data=numeric_columns, palette='magma')
plt.title('Boxplot for Numerical Columns')
plt.show()
```

The Plot shows Outliers in 2 columns i.e.. Age and Fare.Removing outliers in the 'Fare' column ensures a more accurate representation of the dataset's distribution and statistics, especially given that outliers in 'Fare' might significantly impact the analysis compared to outliers in 'Age'.

```python
# Selecting specific columns
columns = ['Age', 'Fare']

# Calculating quartiles and IQR for each column
for col in columns:
    q1 = np.percentile(titanic[col], 25)
    q3 = np.percentile(titanic[col], 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Filtering outliers for each column
    titanic = titanic[(titanic[col] >= lower_bound) & (titanic[col] <= upper_bound)]

# Creating a boxplot after removing outliers
sns.boxplot(data=titanic, palette='magma')
plt.title('Boxplot for dataset after removing outliers')
plt.show()
```

## Modeling

```python
X=titanic[['Sex', 'Pclass']]
Y=titanic['Survived']
```

Split data into test and train by using Sklearn library

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=0)
```

## Create Training Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix
log = LogisticRegression(random_state = 0)
log.fit(X_train, Y_train)
```

## Create Prediction model

```python
pred = log.predict(X_test)
pred
```

```python
print("Accuracy_score :", accuracy_score(Y_test, pred))
print("Matrix :",confusion_matrix(Y_test,pred))
```

```python
from sklearn.metrics import classification_report
print("Classification Report:\n", classification_report(Y_test, pred))
```

```python
Y_test
```

```python
submission=X.iloc[:,:].values
y_final=log.predict(submission)
```

## CONCLUSION

In summary, the logistic regression model applied to the Titanic dataset demonstrated a commendable performance with the following key findings:

- **Accuracy Score:** The model accurately predicted the survival status for approximately 82.64% of passengers in the test set, showcasing a strong overall predictive capability.

- **Confusion Matrix:** The confusion matrix revealed a balanced performance, with 78 true negatives (correctly predicted non-survivors) and 41 true positives (correctly predicted survivors). There were 13 false positives and 12 false negatives, indicating a relatively balanced trade-off between Type I and Type II errors.

- **Precision and Recall:** The precision for predicting non-survival (class 0) was 87%, emphasizing the accuracy of negative predictions. The recall for predicting survival (class 1) was 77%, indicating a satisfactory ability to capture actual instances of positive cases.

- **F1-Score:** The F1-score, which harmonizes precision and recall, was 86% for class 0 and 77% for class 1. This suggests a balanced performance in achieving both precision and recall objectives.

In conclusion, the logistic regression model demonstrated a well-balanced performance in predicting survival outcomes for Titanic passengers.
